```python
# 금천구 스타벅스만 샘플링

df1 = df[(df['gu_name']=='금천구')]

df1
```

```python
# 유클리드 거리

dist_euclidean = distance.cdist(df1[['latitude', 'longitude']], df1[['latitude', 'longitude']],
                                metric= "euclidean")

# 칼럼명, 인덱스 설정

dist_euclidean = pd.DataFrame(data = dist_euclidean, columns= df1['name'])
dist_euclidean.set_index(df1['name'], inplace= True)

dist_euclidean
```

```python
# 맨해튼 거리

dist_manhattan = distance.cdist(df1[['latitude', 'longitude']], df1[['latitude', 'longitude']],
                                metric= "cityblock")

# 칼럼명, 인덱스 설정

dist_manhattan = pd.DataFrame(data = dist_manhattan, columns= df1['name'])
dist_manhattan.set_index(df1['name'], inplace= True)

dist_manhattan.min(numeric_only=True).min(2)
```
